## **Load and Format 2020 Median Home Value**

```python
import pandas as pd
df=pd.read_csv('/content/2020BgHomeValue.csv')
```

```python
df.info()
```

```python

# Convert "2020 Median Home Value" to numeric, forcing non-convertible values to NaN
df['2020 Median Home Value'] = pd.to_numeric(df['2020 Median Home Value'], errors='coerce')

# Remove rows where "2020 Median Home Value" is null after conversion
df_clean = df.dropna(subset=['2020 Median Home Value'])

# Compute the median home value for each facility
facility_median_values = df_clean.groupby('Facility I')['2020 Median Home Value'].median().reset_index()

# Rename the columns for clarity if necessary (optional)
facility_median_values.rename(columns={'2020 Median Home Value': 'MedianHomeValuePerFacility'}, inplace=True)

# Save to CSV
csv_file_path = 'facility_median_values.csv'  # Adjust path if running in a specific environment like Google Colab
facility_median_values.to_csv(csv_file_path, index=False)

# For Google Colab, use the following to download the file:
# from google.colab import files
# files.download(csv_file_path)
```

# **Summary**

```python
import pandas as pd

# Assuming df is your dataframe
# Apply the conditions to filter the dataframe
filtered_df = df[(df['2020 Block Group Size'] < 4576860) &
                 (df['Block Group Size Change'].abs() <= 0.10)]

# Display summary statistics for the specified columns, including the new features
summary = filtered_df[['2020 Median Home Value', '2010 Median Home Value', '2020 Median Rent',
                       '2010 Median Rent', '% HV Change', '% R Change']].describe()

# Apply different formatting to specific summary statistics
formatted_summary = summary.copy()
formatted_summary['% HV Change'] = summary['% HV Change'].map('{:.3f}'.format)
formatted_summary['% R Change'] = summary['% R Change'].map('{:.3f}'.format)
formatted_summary.iloc[:, :4] = formatted_summary.iloc[:, :4].applymap('{:.0f}'.format)

# Display the formatted summary statistics
formatted_summary
```

```python
# Assuming 'df' is your original DataFrame and 'filtered_df' is the filtered DataFrame

# First, find the rows in the original DataFrame that are not in the filtered DataFrame
filtered_out_df = df.merge(filtered_df, indicator=True, how='outer').query('_merge=="left_only"').drop('_merge', axis=1)

# Export the filtered out data to a CSV file
filtered_out_df.to_csv('/content/filtered_out_data2.csv', index=False)

# Use Google Colab's files.download to download the CSV file to your local machine
from google.colab import files
files.download('/content/filtered_out_data2.csv')
```

# **Analyzing Exclusions**

```python
import pandas as pd

# Total number of rows in the original dataset
total_rows = len(df)

# Number of rows meeting the first condition (2020 Block Group Size < 4576860)
rows_condition1 = len(df[df['2020 Block Group Size'] < 4576860])

# Number of rows meeting the second condition (Block Group Size Change within -10% to +10%)
rows_condition2 = len(df[df['Block Group Size Change'].abs() <= 0.10])

# Calculate the number of rows excluded by each condition
excluded_rows_condition1 = total_rows - rows_condition1
excluded_rows_condition2 = total_rows - rows_condition2

# Print the results
print(f"Rows excluded by the '2020 Block Group Size' condition: {excluded_rows_condition1}")
print(f"Rows excluded by the 'Block Group Size Change' condition: {excluded_rows_condition2}")
```

# **2020 Housing Price and Rent Correlations**

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Assuming filtered_df is your filtered DataFrame and it has columns '2020 Median Home Value' and '2020 Median Rent'

# Normalize the data from the filtered dataset
scaler = MinMaxScaler()
normalized_data = scaler.fit_transform(filtered_df[['2020 Median Home Value', '2020 Median Rent']])
normalized_df = pd.DataFrame(normalized_data, columns=['2020 Median Home Value', '2020 Median Rent'])

# Compute correlation coefficient for the filtered dataset
correlation_coef = normalized_df['2020 Median Home Value'].corr(normalized_df['2020 Median Rent'])

# Plotting
plt.figure(figsize=(10, 6))
sns.regplot(x='2020 Median Home Value', y='2020 Median Rent', data=normalized_df, line_kws={"color": "red"})
plt.title(f'Scatter Plot of Normalized Data with Regression Line\nCorrelation Coefficient: {correlation_coef:.3f}')
plt.xlabel('2020 Median Home Value (Normalized)')
plt.ylabel('2020 Median Rent (Normalized)')
plt.grid(True)
plt.show()
```

# **Baseline Model with Normalized Target Variable**

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import OneHotEncoder, StandardScaler

# Load the dataset
data = pd.read_csv('/content/modified_file Systems_Indepdendents_regression.csv')

# Prepare the features and target variable, excluding the specified columns
X = data.drop(['Facility ID', '2020_Median_Home_Value'], axis=1)
y = data['2020_Median_Home_Value']

# Identify categorical columns for one-hot encoding
categorical_cols = ['Near_Highway', 'Cluster', 'Primary Mode', 'Facility_Type 'System', 'Decade Built']

# OneHotEncoder instance to transform categorical variables
encoder = OneHotEncoder(sparse=False, handle_unknown='ignore')
X_encoded = encoder.fit_transform(X[categorical_cols])

# Create a new DataFrame from the encoded variables
encoded_df = pd.DataFrame(X_encoded, columns=encoder.get_feature_names_out(categorical_cols))

# Combine the encoded DataFrame with the original data (excluding targets and categorical columns)
X_preprocessed = pd.concat([X.drop(categorical_cols, axis=1), encoded_df], axis=1)

# Fill missing values for numerical features with their median
X_preprocessed = X_preprocessed.fillna(X_preprocessed.median())

# Removing rows with missing values in the target variable '2020 Median Home Value'
data_cleaned = data.dropna(subset=['2020 Median Home Value'])
y_cleaned = data_cleaned['2020 Median Home Value']
X_cleaned = X_preprocessed.loc[data_cleaned.index, :]

# Standardize the target variable
scaler_y = StandardScaler()
y_cleaned_scaled = scaler_y.fit_transform(y_cleaned.values.reshape(-1, 1)).flatten()

# Split the scaled data into training and testing sets
X_train_cleaned, X_test_cleaned, y_train_cleaned_scaled, y_test_cleaned_scaled = train_test_split(
    X_cleaned, y_cleaned_scaled, test_size=0.2, random_state=42)

# Initialize and fit the Linear Regression model on the scaled target
lin_reg_cleaned = LinearRegression()
lin_reg_cleaned.fit(X_train_cleaned, y_train_cleaned_scaled)

# Make predictions on the scaled target
y_pred_cleaned_scaled = lin_reg_cleaned.predict(X_test_cleaned)

# Evaluate the model on the scaled target
r2_cleaned_scaled = r2_score(y_test_cleaned_scaled, y_pred_cleaned_scaled)
mse_cleaned_scaled = mean_squared_error(y_test_cleaned_scaled, y_pred_cleaned_scaled)

# Inverse transform the predictions to get them back on the original scale for interpretability
y_pred_cleaned = scaler_y.inverse_transform(y_pred_cleaned_scaled.reshape(-1, 1)).flatten()

# Evaluate the inverse-transformed predictions
r2_cleaned = r2_score(y_test_cleaned, y_pred_cleaned)
mse_cleaned = mean_squared_error(y_test_cleaned, y_pred_cleaned)

print(f'R-squared score: {r2_cleaned_scaled}')
print(f'Mean Squared Error: {mse_cleaned_scaled}')

```
