```python
import geopandas as gpd
import folium
from shapely.geometry import Point
import pandas as pd

def check_amenities_nearby(lat, lon, radius_meters, tags):
    api = overpy.Overpass()
    results_dict = {}

    for tag in tags:
        key, value = tag.split("=")
        query = f"""
        [out:json];
        (
            node["{key}"="{value}"](around:{radius_meters},{lat},{lon});
            way["{key}"="{value}"](around:{radius_meters},{lat},{lon});
        );
        out body;
        """
        # Execute the query
        result = api.query(query)
        # Count the presence of the amenity/tag
        results_dict[tag] = len(result.nodes) + len(result.ways)

    return results_dict

# Read the longitude and latitude coordinates from the file
coordinates_df = pd.read_csv('/content/MTA_NYCT1.csv')  # Replace with your file path
facility_id = input("Enter the Facility ID: ")  # User inputs the Facility ID

# Filter the DataFrame for the selected Facility ID
selected_location = coordinates_df[coordinates_df['Facility ID'] == int(facility_id)].iloc[0]
lat, lon = selected_location['Latitude'], selected_location['Longitude']

# Read the TIGER/Line Shapefile
block_groups = gpd.read_file('/content/tl_2020_36_tabblock20 (1).zip')

# Define the point
fixed_point = Point(lon, lat)

# Create a buffer around the point
fixed_point_series = gpd.GeoSeries([fixed_point], crs="EPSG:4326")
fixed_point_series = fixed_point_series.to_crs("EPSG:32618")  # Reproject to UTM CRS
fixed_point_buffer = fixed_point_series.buffer(402)[0]  # Buffer in meters
fixed_point_buffer = gpd.GeoSeries([fixed_point_buffer], crs="EPSG:32618").to_crs("EPSG:4326")[0]

# Select Census Block Groups within 1/4 mile of the coordinates
block_groups_nearby = block_groups[block_groups.intersects(fixed_point_buffer)]

# Convert GeoDataFrames to GeoJSON
block_groups_geojson = block_groups_nearby.to_json()

# Create a folium map centered around your coordinates
m = folium.Map(location=[lat, lon], zoom_start=15)

# Add the base street map
folium.TileLayer('OpenStreetMap').add_to(m)

# Add the census block groups layer
folium.GeoJson(block_groups_geojson, name='Census Block Groups').add_to(m)

# Add the buffer layer
folium.GeoJson(fixed_point_buffer, name='Buffer Zone', style_function=lambda x: {'fillColor': 'green', 'color': 'green', 'fillOpacity': 0.3}).add_to(m)

# Add a marker for the original point
folium.Marker([lat, lon], popup='Fixed Point').add_to(m)

# Add layer control to toggle layers
folium.LayerControl().add_to(m)

# Display the map
m
```
