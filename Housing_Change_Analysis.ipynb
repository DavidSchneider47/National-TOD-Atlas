# **Transit Systems with the Greatest Housing Development Around Stations**

```python
import pandas as pd

# Replace 'yourfile.csv' with the path to your CSV file
file_path = '/content/FINAL_Clusters_TOD_Score_Pop_Change_Indepdenentsv1.csv'

# Read the CSV file
data = pd.read_csv(file_path)

# Ensure numeric conversion for housing data
data['2020 Housing Units'] = pd.to_numeric(data['2020 Housing Units'], errors='coerce')
data['2010 Housing Units'] = pd.to_numeric(data['2010 Housing Units'], errors='coerce')

# Remove any transit agency with a 2010 housing unit count of 0
data = data[data['2010 Housing Units'] > 0]

# Calculate averages for each transit agency and count of Facility IDs
agency_averages = data.groupby('Transit Agency Name').agg(
    Average_2020_Housing_Units=('2020 Housing Units', 'mean'),
    Average_2010_Housing_Units=('2010 Housing Units', 'mean'),
    Facility_Count=('Facility ID', 'count')  # Count the number of Facility IDs for each agency
).reset_index()

# Calculate difference and percentage change, then round to no decimal places
agency_averages['Difference'] = (agency_averages['Average_2020_Housing_Units'] - agency_averages['Average_2010_Housing_Units']).round(0)
agency_averages['Percentage Change'] = (((agency_averages['Difference'] / agency_averages['Average_2010_Housing_Units']) * 100).round(0))

# Ensure the average housing units are also rounded to no decimal places
agency_averages['Average_2020_Housing_Units'] = agency_averages['Average_2020_Housing_Units'].round(0)
agency_averages['Average_2010_Housing_Units'] = agency_averages['Average_2010_Housing_Units'].round(0)

# Sort by percentage change in descending order
sorted_agencies = agency_averages.sort_values(by='Percentage Change', ascending=False)

# Select the top 10 transit agencies based on percentage change
top_agencies = sorted_agencies.head(20)

# Display the results
top_agencies[['Transit Agency Name', 'Average_2020_Housing_Units', 'Average_2010_Housing_Units', 'Difference', 'Percentage Change', 'Facility_Count']]
```

# **Housing Availability by TOD Market**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Replace 'yourfile.csv' with the path to your CSV file
file_path = '/content/FINAL_Clusters_TOD_Score_Pop_Change_Indepdenentsv1.csv'

# Read the CSV file
data = pd.read_csv(file_path)

# Ensure numeric conversion for housing units data
data['2020 Housing Units'] = pd.to_numeric(data['2020 Housing Units'], errors='coerce')

# Remove rows where '2020 Housing Units' are NaN or zero to avoid division errors
filtered_data = data[data['2020 Housing Units'] > 0]

# Calculate total housing units in 2020
total_housing_units_2020 = filtered_data['2020 Housing Units'].sum()

# Calculate the percentage of total housing units in 2020 for each cluster
filtered_data['2020 Housing Units % of Total'] = (filtered_data['2020 Housing Units'] / total_housing_units_2020) * 100
cluster_housing_percentage = filtered_data.groupby('Cluster Name')['2020 Housing Units % of Total'].sum().reset_index()

# Plotting
plt.figure(figsize=(10, 6))
bars = plt.bar(cluster_housing_percentage['Cluster Name'], cluster_housing_percentage['2020 Housing Units % of Total'], color='blue')
plt.xlabel('Cluster Name')
plt.ylabel('Percentage of Total Housing Units (%)')
plt.title('Percentage of Total 2020 Housing Units by Cluster')
plt.xticks(rotation=45)

# Adding the percentage on top of each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval, f'{yval:.1f}%', ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

## **Backward Propigation to Identify the Most Salient Features**

```python
import statsmodels.api as sm

# Define the backward elimination function
def backward_elimination(X, y, significance_level=0.05):
    num_vars = len(X.columns)
    for i in range(0, num_vars):
        regressor_OLS = sm.OLS(y, sm.add_constant(X)).fit()
        max_p_value = max(regressor_OLS.pvalues.iloc[1:])  # Skip the constant's p-value
        if max_p_value > significance_level:
            feature_with_max_pvalue = regressor_OLS.pvalues[regressor_OLS.pvalues == max_p_value].index[0]
            X = X.drop(feature_with_max_pvalue, axis=1)
        else:
            break
    return X.columns.tolist()

# Assuming 'X_encoded' is your dataset with all encoded features and 'y' is your target variable
significant_features = backward_elimination(X_encoded, y)

# Print the significant features
print("Significant features after backward elimination:", significant_features)
```
